#### <font color = 'gray'>I  will be using the below function to train_test_split,fit ,predict my model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,ElasticNet, Ridge
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor,StackingRegressor,GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import Lasso
from sklearn.ensemble import VotingRegressor
import xgboost as xgb
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.ensemble import ExtraTreesRegressor
```

```python
def model_training(data, target_col, model, test_size, seed):
    '''This function runs the test_train_split process on a transformed data
    it accepts 4 inputs namely
    data: the transformed data we wantr to test_train_split
    model: this model must have been fitted and instatiated and assigned a variable name
    test_size: depends on the size of the transformed data, it increases with the size of data,represents the amount of the dataset we put aside to test on our model
    target column : the predicted column
    seed: a number to ensure uniformity of answers across board usually 42'''
    
    X = data.drop(columns=[target_col])
    y = data[target_col]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
    
    model.fit(X_train, y_train)
    
    y_pred = model.predict(X_test)
    
    
    
    return  X_train, X_test, y_train, y_test,y_pred,model                  
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,r2_score
```

```python
def reg_model_evaluation(model_name,y_test,y_pred):
    ''' This function accepts two  inputs namely:
    - model_name : the name of the model to be evaluated
    - y-pred : the predicted value of my actual test
    - y-test: the  actual test
    but outputs a print statement of 
    1. mean_squared_error
    2. root_mean_squared_error
    3. mean_absolute_error
    4. mean_absolute_percentage_error 
    5. r_score  '''
    
    mse = mean_squared_error(y_test,y_pred)
    rmse = np.sqrt(mse)
    mae =mean_absolute_error(y_test,y_pred)
    mape = mean_absolute_percentage_error(y_test,y_pred)
    r2_ = r2_score(y_test,y_pred)
    print(f'The evaluation below is for {model_name} model')
    print(f'The mean_squared_error is {round(mse,2)}')
    print(f'The root_mean_squared_error is {round(rmse,2)}')
    print(f'The mean_absolute_error is {round(mae,2)}')
    print(f' mean_absolute_percentage_error is {round(mape,2)}')
    print(f'The cofficient_of_determination(r2) is {round(r2_,2)*100}% ')
```
